```python
from fer import FER
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/content/image.jpg')
detector = FER()
emotion= detector.detect_emotions(img)
print(emotion)
```

```python
(x, y, w, h) = emotion[0]["box"]  # Get the bounding box coordinates

# Find the top emotion manually
#emotions = face["emotions"]
emotion, score = detector.top_emotion(img)  # Get the top emotion

# Draw a rectangle around the face
cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Put the emotion label with the probability score
text = f"{emotion} ({score:.2f})"
cv2.putText(img, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

# Display the image with the emotion labels
plt.imshow(img)
plt.axis('off')
plt.show()
```
